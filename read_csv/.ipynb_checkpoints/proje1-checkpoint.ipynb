```python
#min-max normalizasyonu
def MinMaxNormalizasyonu(data):
    minVeri=np.min(data)
    maxVeri=np.max(data)
    sonuc=(data-minVeri) / (maxVeri-minVeri)
    return sonuc
```

```python
normalizeEdilmişData=MinMaxNormalizasyonu(data)
print("Normalizasyon öncesi:",data)
print("*************************************************************")
print("Normalizasyon sonrası:", normalizeEdilmişData)
```

```python
#İstatiksel normalizasyon
def istatikselNormalizasyon(data):
    ortalama=np.mean(data)
    standartSapma=np.std(data)
    sonuc=(data-ortalama) /standartSapma
    return sonuc
```

```python
normalizeEdilmişData2=istatikselNormalizasyon(data)
print("Normalizasyon öncesi:",data)
print("*************************************************************")
print("Normalizasyon sonrası:", normalizeEdilmişData2)
```
