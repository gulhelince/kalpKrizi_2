```python
#Eksik veri yok o halde kategorik veri kontrolü yapılabilir.
#Sütunların kategorik mi yoksa sayısal mı olduğu kontrol edilir.
for i in list(data.columns): #dataya dönüştürmesini sağlar 
    print("{} -- {}".format(i,data[i].value_counts().shape[0]))
```

```python
#Kategorik ve sayısal değerler birbirine katmadan öğrenme işlemi yapmayı sağlar 
kategorik=["sex","cp","fbs","restecg","exng","slp","caa","thall","output"]
```

```python
data_kat=data.loc[:,kategorik] #Dataframe oluştur
```

```python
data_kat
```

```python
for i in data_kat: #tüm sütunlarda görselleştirmeyi sağlar 
    sns.countplot(x=i,data=data_kat,hue="output")
    # x=her biri ayrı ayrı gelen değerlerden oluşur 
    # data yani hangi verinin alınacağı belirtilmelidir.
    # hue=neye göre oluşturacağı girilmelidir.
    plt.show()
```
