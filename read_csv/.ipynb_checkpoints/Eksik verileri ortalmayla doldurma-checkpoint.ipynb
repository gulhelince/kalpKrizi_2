# Eksik verileri ortalmayla doldurma 

```python
from sklearn.impute import SimpleImputer
eksikVeriTamamlama=SimpleImputer(missing_values=np.nan,strategy='mean')
eksikVeriTamamlama=eksikVeriTamamlama.fit(veri.iloc[:,1:4])
veri.iloc[:,1:4]=eksikVeriTamamlama.transform(veri.iloc[:,1:4])
```

```python
veri
```

```python
veri=pd.read_csv("eksikveriler .csv")

from sklearn.impute import SimpleImputer
eksikVeriTamamlama=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
eksikVeriTamamlama=eksikVeriTamamlama.fit(veri.iloc[:,1:4])
veri.iloc[:,1:4]=eksikVeriTamamlama.transform(veri.iloc[:,1:4])
```

```python
veri #eksik veriler en çok tekrar eden değerle doldurulur 
```

```python
veri=pd.read_csv("eksikveriler .csv")

from sklearn.impute import SimpleImputer
eksikVeriTamamlama=SimpleImputer(missing_values=np.nan,strategy='median')
eksikVeriTamamlama=eksikVeriTamamlama.fit(veri.iloc[:,1:4])
veri.iloc[:,1:4]=eksikVeriTamamlama.transform(veri.iloc[:,1:4])
```
