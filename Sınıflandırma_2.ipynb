```python
data=pd.read_csv("data.csv") #datayı al
data.drop(["Unnamed: 32"],axis=1,inplace=True) #datayı sil ve dataya uygula
data.drop(["id"],axis=1,inplace=True)

data.diagnosis.replace({"M":1, "B":0},inplace=True)
y=data.diagnosis.values
X=data.drop(["diagnosis"],axis=1) #sınıf hariç data

#datayı eğitim ve test olarak ayır
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=44)

knn=KNeighborsClassifier(n_neighbors=7) #Yapıcı metodu çağrılır
knn.fit(X_train,y_train) #metotun çalıştığını gösterir

tahmin=knn.predict(X_test)  

sonuc=knn.score(X_test,y_test)
print("model başarısı:",sonuc)
```

```python
k_değerleri=[]
for i in range(1,29,2):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    k_değerleri.append(knn.score(X_test,y_test))

    
```
