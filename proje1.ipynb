```python
#min-max normalizasyonu
def MinMaxNormalizasyonu(data):
    minVeri=np.min(data)
    maxVeri=np.max(data)
    sonuc=(data-minVeri) / (maxVeri-minVeri)
    return sonuc
```

```python
normalizeEdilmiData=MinMaxNormalizasyonu(data)
print("Normalizasyon ncesi:",data)
print("*************************************************************")
print("Normalizasyon sonras:", normalizeEdilmiData)
```

```python
#statiksel normalizasyon
def istatikselNormalizasyon(data):
    ortalama=np.mean(data)
    standartSapma=np.std(data)
    sonuc=(data-ortalama) /standartSapma
    return sonuc
```

```python
normalizeEdilmiData2=istatikselNormalizasyon(data)
print("Normalizasyon ncesi:",data)
print("*************************************************************")
print("Normalizasyon sonras:", normalizeEdilmiData2)
```
