```python
data1=data.copy() # datayı kopyala
#datada normalizasyon işlemi yapılmaz ise hata oranı artar.
#
```

```python
data1.head()
```

```python
data1=pd.get_dummies(data1,columns=kategorik[:-1],drop_first=True)
# drop_first=True , eski değerleri atmasını ister 
data1.head()
```

```python
X=data1.drop(["output"],axis=1)
```

```python
X
```

```python
y=data1[["output"]]
```

```python
y
```

```python
#normalizasyon işlemi yapılır
scaler=StandardScaler()
```

```python
X[sayısal[:-1]]=scaler.fit_transform(X[sayısal[:-1]])
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=44)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier() #Yapıcı metodu çağrılır
knn.fit(X_train,y_train) #metotun çalıştığını gösterir
```

```python
tahmin=knn.predict(X_test)  
```

```python
tahmin
```

```python
y_test #gerçek değerler 
```

```python
sonuc=knn.score(X_test,y_test)
print("model başarısı:",sonuc)
```

```python
k_değerleri=[]
for i in range(1,20,2):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    k_değerleri.append(knn.score(X_test,y_test))
```

```python
k_değerleri
```

```python
#Karışık bir veri seti üzerinde kategorik veri setiyle sayısal veri setini birbirine karıştırmadan yapılan işlemlerdir .
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.15,random_state=44)
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(Xtrain,ytrain)
```

```python
y_prede=logreg.predict(Xtest)
y_prede
```

```python
y_pred=logreg.predict_proba(Xtest)
y_pred
```

```python
Xtest
```

```python
print("Test accuracy",accuracy_score(y_prede,ytest))
```
