```python
import numpy as np
```

```python
data=np.random.rand(5)*100
```

```python
data
```

```python
def MinMaxNormalizyonu(data):
    minVeri=np.min(data)
    maxVeri=np.max(data)
    sonuc=(data-minVeri) /(maxVeri-minVeri)
    return sonuc
```

```python
normalizeEdilmişData=MinMaxNormalizyonu(data)
print("Normalizasyon öncesi:", data)
print("*************************************************************")
print("Normalizasyon sonrası:", normalizeEdilmişData)
```

```python
def İstatistikselNormalizasyon(data):
    ortalama=np.mean(data)
    standartSapma=np.std(data)
    sonuc=(data-ortalama) / (standartSapma)
    return sonuc
```

```python
normalizeEdilmişData2=İstatistikselNormalizasyon(data)
print("Normalizasyon öncesi:", data)
print("*************************************************************")
print("Normalizasyon sonrası:", normalizeEdilmişData2)
```

```python
def logaritmikNormalizasyon(data):
    return np.log(data)
```

```python
normalizeEdilmişData3=logaritmikNormalizasyon(data)
print("Normalizasyon öncesi:", data)
print("*************************************************************")
print("Normalizasyon sonrası:", normalizeEdilmişData3)
```

```python
def KarakokNormalizasyonu(data):
    return np.sqrt(data)
```

```python
normalizeEdilmişData4=KarakokNormalizasyonu(data)
print("Normalizasyon öncesi:", data)
print("*************************************************************")
print("Normalizasyon sonrası:", normalizeEdilmişData4)
```
